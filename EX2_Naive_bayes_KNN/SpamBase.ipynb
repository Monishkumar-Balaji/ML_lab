# Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler, RobustScaler
```

```python
num_colmn = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
num_colmn = num_colmn.drop('class')
df[num_colmn] = MinMaxScaler().fit_transform(df[num_colmn])
```

```python
df
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB,BernoulliNB
from sklearn.metrics import (
accuracy_score, precision_score, recall_score, f1_score,
confusion_matrix, classification_report, roc_auc_score,
roc_curve, average_precision_score
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_drp = df.drop('class',axis=1)
y = df['class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_drp,y,random_state=42, test_size= 0.2)
```

```python
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

# MultinomialNB

```python
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
```

```python
mnbpred = mnb.predict(X_test)
```

```python
print(classification_report(y_test,mnbpred))
```

# Bernoulli Naive Bayes

```python
bnb = BernoulliNB()
bnb.fit(X_train,y_train)
```

```python
bnbpred = bnb.predict(X_test)
```

```python
print(classification_report(y_test,bnbpred))
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
import time
```

```python
knn = KNeighborsClassifier()
x = time.time()
knn.fit(X_train,y_train)
y = time.time()
```

```python
print(y-x)
knnpred = knn.predict(X_test)
```

```python
print(classification_report(y_test,knnpred))
```

# KNeighborsClassifier - Ball_tree Algorithm

```python
knn_balltree = KNeighborsClassifier(algorithm = 'ball_tree')
x = time.time()
knn_balltree.fit(X_train,y_train)
y = time.time()
```

```python
print(y-x)
knn_balltree_pred = knn_balltree.predict(X_test)
```

```python
print(classification_report(y_test,knn_balltree_pred))
```

# KNeighborsClassifier - Kd_tree Algorithm

```python
knn_kdtree = KNeighborsClassifier(algorithm = 'kd_tree')
x = time.time()
knn_kdtree.fit(X_train,y_train)
y = time.time()
```

```python
print(y-x)
knn_kdtree_pred = knn_kdtree.predict(X_test)
```

```python
print(classification_report(y_test,knn_kdtree_pred))
```

# Hyperparameters Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5,shuffle= True, random_state= 42)
```

```python
para = {
    'n_neighbors': [1,3,5,7,9,11,15],
    'p':[1,2],
    'weights':['uniform','distance']
}

grid = GridSearchCV(KNeighborsClassifier(),param_grid= para, cv = skf, scoring = 'accuracy')

```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
print(classification_report(y_test,grid.predict(X_test)))
```
